### Faça um gráfico do tipo "scatter" do arquivo "stars.txt" presente na seção 3.2 de nosso livro texto.


```python
from pylab import scatter,xlabel,ylabel,xlim,ylim,show
from numpy import loadtxt
data = loadtxt("C:/Users/angel/Documents/FísicaComputacional-2022,q3/semana3/cpresources/stars.txt",float)
x = data[:,0]
y = data[:,1]
scatter(x,y)
xlabel("Temperature")
ylabel("Magnitude")
xlim(0,13000)
ylim(-5,20)
show()
```
