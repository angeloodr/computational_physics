```python
import matplotlib.pyplot as plt

data = []
with open("C:/Users/angel/Documents/FísicaComputacional-2022,q3/semana3/cpresources/sunspots.txt") as f:
    for lines in f:
        data.append(lines.split('\t'))
for i, linha in enumerate(data):
    data[i][1] = data[i][1][:-1]
    for j, item in enumerate(linha):
        data[i][j] = float(data[i][j])
       
tempo, posicao = [], []
for linha in data[:]:
    tempo.append(linha[0])
    posicao.append(linha[1])
plt.figure()
plt.plot(tempo, posicao)
plt.xlabel("tempo")
plt.ylabel("sunspots")
```

```python
plt.figure()
plt.plot(tempo[:1000], posicao[:1000])
plt.xlabel("tempo")
plt.ylabel("sunspots")
```

```python
# runnig average
def runAvg(data, r=5):
    runaverage = []
    avaliado = data[5:-5]
    for m, item in enumerate(avaliado):
        datasum = 0
        m = m+5 #pq temos que ver o dado no data que vai estar 5 posições a mais de diferenca com o avaliado
        for i in range(m-r, m+r):
            datasum += data[i]
        runaverage.append( (1/(2*r) * datasum) )
    return runaverage

runningaverage = runAvg(posicao)
plt.figure()
plt.plot(tempo[:1000], posicao[:1000],'y', label = "bruto")  
plt.plot(tempo[5:1005], runningaverage[:1000], 'k', label="runningAverage")
plt.legend()
plt.xlabel("tempo")
plt.ylabel("sunspots")  
```
