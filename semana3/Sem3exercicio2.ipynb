tendo a equação $ \dot x = rx(1-x)$, sendo que $r$ é uma constante.

passo 1: Comece com o valor de $x=\frac 1 2$

passo 2: itere a equação diferencial 1000 vezes de forma que o valor de x não mude mais de forma drastica

passo 3: Guarde o ultimo valor de x encontrado 

passo 4: faça de novo os passos 1, 2, 3 variando o $r$ de 1 a 4 em passos de 0.01

passo 5: plot x(r)

passo 6: responda as questões 

    6.1 For a given value of r what would a fixed point look like on the Feigenbaum plot? How about a limit cycle? And what would chaos look like?

    6.2 Based on your plot, at what value of r does the system move from orderly behavior (fixed points or limit cycles) to chaotic behavior? This point is sometimes called the “edge of chaos.”

```python
import matplotlib.pyplot as plt
import numpy as np
 
def chaosEq(x, r):
    return r*x*(1-x)

rvec = np.arange(1,4,0.01)
x_0 = 1/2
niteracao = 1000
x_final = []

for r in rvec:
    x = x_0
    for passo in range(niteracao):
        x = chaosEq(x, r)
    x_final.append(x)

plt.figure()
plt.plot(rvec, x_final, 'ok')
plt.title("Feigenbaum plot")
plt.xlabel("r")
plt.ylabel('x')
```

6.1 For a given value of r what would a fixed point look like on the Feigenbaum plot? How about a limit cycle? And what would chaos look like?

-> O ponto fixo seria os pontos proximos ao valor de $0.5$. O "limit cycle" seriam os pontos que estão ordenados e que serguiriam uma função mais bem definida no gráfico. Já os pontos caóticos estariam espalhados sem seguir nenhum padrão claro

6.2 Based on your plot, at what value of r does the system move from orderly behavior (fixed points or limit cycles) to chaotic behavior? This point is sometimes called the “edge of chaos.”
