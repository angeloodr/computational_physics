```python
def somaDireta(N):
    soma = 0
    for n in range(1,2*N+1):
        soma += ( (-1)**n *n ) /(n+1)
    return soma

def somaDividida(N):
    soma1 = 0
    soma2 = 0
    for n in range(1,N+1):
        soma1 += (2*n-1)/(2*n)
        soma2 += (2*n)/(2*n+1)
    return (-1)*soma1 + soma2

def somaSimplificada(N):
    soma = 0
    for n in range(1,2*N+1):
        soma += 1/(2*n*(2*n+1))
    return soma

N=10
print(somaDireta(N),somaDividida(N), somaSimplificada(N))
```

```python
import numpy as np
import matplotlib.pyplot as plt
N=10**4
eixox = np.log10(np.arange(N))
erroRel1, erroRel2 = [], []
for n in range (N): 
    erroRel1.append((somaDireta(n)-somaSimplificada(N))/somaSimplificada(N))
    erroRel2.append((somaDividida(N)-somaSimplificada(N))/somaSimplificada(N))
plt.figure()
plt.plot(eixox, erroRel1)
plt.plot(eixox, erroRel2)
```
